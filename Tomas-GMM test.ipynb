```python
from sac.distributions import GMM
import tensorflow as tf
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
action_dim = 2
with tf.variable_scope("", reuse=tf.AUTO_REUSE):
    conditional = tf.placeholder(shape=(None,2), dtype="float32")
    gmm = GMM(K=2, Dx=action_dim, cond_t_lst=[conditional])
```

```python
num_obs = 1000
obs = np.zeros((num_obs, 2))
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
x, logp = sess.run([gmm.x_t, gmm.log_p_t], feed_dict = {conditional: obs})
```
